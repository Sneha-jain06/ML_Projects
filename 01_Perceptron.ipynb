```python
xTrain = [i for i in range(21)]
yTrain= [i for i in range(10*20+1) if i%10==0]
```

```python
print(yTrain)
```

```python
def tempFunc(x):
  y=10*x
  return y
for i in xTrain:
  print(tempFunc(i))
```

Approach 2

```python
x= [i for i in range(21)]
y= [i for i in range(10*20+1) if i%10==0]

print(x)
print(y)
```

```python
xTrain  = x[:-5]
yTrain  = y[:-5]

xTest = x[-5:]
yTest = y[-5:]

print(f'this is xTrain {xTrain}')
print(f'this is yTrain {yTrain}')
print(f'this is xTest { xTest}')
print(f'this is xTrain {yTest}')
```

```python
import tensorflow as tf
from tensorflow import keras 

model = tf.keras.Sequential([tf.keras.layers.Dense(units=1, input_shape=[1])])
```

```python
model.compile(optimizer='adam',loss='mae') #optimizers --> sgd, rmsprop, adamax, 
# adagrad ; loss --> mse, mae (this is regression sample
```

```python
history =model.fit(x=xTrain, y=yTrain ,validation_data=(xTest,yTest),epochs=5000)
```
